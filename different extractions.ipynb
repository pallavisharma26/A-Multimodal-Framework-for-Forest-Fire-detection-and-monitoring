```python
from skimage import io
from matplotlib import pyplot as plt
from skimage.filters import roberts, prewitt, sobel, scharr
```

```python
image = io.imread("D:\\forest fire dataset\\final code\\fire.1.png",as_gray=True)
```

```python
edge_roberts=roberts(image)
edge_prewitt=prewitt(image)
edge_sobel=sobel(image)
edge_scharr=scharr(image)
plt.imshow(image,cmap='gray')
```

```python
fig,axes=plt.subplots(nrows=2,ncols=2,sharex=True,sharey=True,figsize=(20,20))
ax=axes.ravel()


ax[0].imshow(edge_roberts,cmap=plt.cm.gray)
ax[0].set_title("Roberts Edge Detection")

ax[1].imshow(edge_sobel,cmap=plt.cm.gray)
ax[1].set_title("Sobel Edge Detection")

ax[2].imshow(edge_scharr,cmap=plt.cm.gray)
ax[2].set_title("Scharr Edge Detection")

ax[3].imshow(edge_prewitt,cmap=plt.cm.gray)
ax[3].set_title("Prewitt Edge Detection")
```

```python
from skimage.feature import canny
```

```python
import numpy as np
from skimage import restoration
```

```python
from skimage.filters.rank import entropy
from skimage.morphology import disk
```

```python
entropy_image=entropy(image,disk(3))
plt.imshow(entropy_image,cmap='gray')
```

```python
from skimage.filters import try_all_threshold
fig,ax=try_all_threshold(entropy_image, figsize=(10,8),verbose=False)
plt.show()
```

```python
from skimage.filters import threshold_otsu
```

```python
thres=threshold_otsu(entropy_image)
binary=entropy_image<=thres
plt.imshow(binary,cmap='gray')
```
